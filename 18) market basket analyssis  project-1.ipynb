# Associated Rule advance

```python
b=pd.read_csv("C:/Users/91973/Downloads/market basket analyssis.csv")
```

```python
b.shape
```

```python
tr=[]
for i in range(0,7500):
    tr.append([str(b.values[i,j]) for j in range(0,20)])
```

```python
tr
```

```python
from apyori import apriori
```

```python
rule=apriori(tr,min_support=0.2,min_left=1)
```

```python
rule
```

```python
r=list(rule)
```

```python
re_list=[]
```

```python
for i in range(0,len(r)):
    re_list.append("rule:"+str(r[i][0])+"append:"+str(r[i][1]))
```

```python
print(re_list)
```
